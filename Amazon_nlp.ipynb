```python
import pandas as pd
```

```python
df=pd.read_csv("Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv")
df.head()
df.tail()
```

```python
df.isnull().sum()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from textblob import TextBlob, Word

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def cleaned_data(text):
    # 1. K√º√ß√ºk harfe √ßevir
    text = text.lower()

    # 2. Noktalama i≈üaretlerini kaldƒ±r
    text = re.sub(r'[^\w\s]', '', text)

    # 3. Tokenize edip lemmatize et
    words = [Word(word).lemmatize() for word in text.split()]

    # 4. Stopwords'leri √ßƒ±kar
    words = [word for word in words if word not in stop_words]

    # 5. Tekrar birle≈ümi≈ü hali
    return ' '.join(words)
```

```python
df['cleaned_text'] = df['reviews.text'].astype(str).apply(cleaned_data)
df[['reviews.text', 'cleaned_text']].sample(5)
```

```python
df = df[df['reviews.rating'].isin([1, 2, 3, 4, 5])]  # sadece ge√ßerli pu
```

```python
#TFƒ∞DF
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(df['reviews.text'])
X
```

```python
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, df['label'], test_size=0.2, random_state=42
)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score

y_prob = model.predict_proba(X_test)[:, 1]
print("AUC Score:", roc_auc_score(y_test, y_prob))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(class_weight='balanced', max_iter=1000)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
y_prob = lr.predict_proba(X_test)[:, 1]

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("AUC Score:", roc_auc_score(y_test, y_prob))
```

```python
#sample
samples = [
    "This product is absolutely amazing. I love it!",
    "Terrible quality. It stopped working in a week.",
    "Very satisfied with the purchase. Would recommend.",
    "Not worth the price. I'm very disappointed.",
    "It works okay, but the battery life is poor."
]
sample_cleaned = [cleaned_data(text) for text in samples]
sample_vector = tfidf.transform(sample_cleaned)
preds = lr.predict(sample_vector)
probs = lr.predict_proba(sample_vector)[:, 1]
for i, sentence in enumerate(samples):
    print(f"üìù C√ºmle: {sentence}")
    print("üîç Temiz:", sample_cleaned[i])
    print("üìä Tahmin:", "Pozitif üòä" if preds[i] == 1 else "Negatif üò†")
    print(f"üìà Pozitif olasƒ±lƒ±ƒüƒ±: {round(probs[i]*100, 2)}%\n")
```
